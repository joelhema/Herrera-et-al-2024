This file contains R functions that reproduce the analysis permormed in the discrete network from **Herrera et al. 2024. A system-Level Model Reveals that Transcriptional Stochasticity is Required for Hematopoietic Stem Cell Differentiation**

```r

# install BoolNet if nedeed 
install.packages('BoolNet')

# Load depedencies
library(BoolNet)

# Load network file in BoolNet style 
discrete_network = loadNetwork('BoolNet_HSC_network.txt')

# Compute attractors using an asynchronous udpate scheme
asynchronous_attractors = getAttractors(discrete_network, type = c('asynchronous') )
plotAttractors(asynchronous_attractors)

# Simulate a p53 Loss-of-Function mutation 
p53_LOF = fixGenes(discrete_network, 'p53',0)
p53_LOF_attractors = getAttractors(p53_LOF, type= c('asynchronous'))
plotAttractors(p53_LOF_attractors)

# Simulate a p53 Gain-of-Function mutation 
p53_GOF = fixGenes(discrete_network, 'p53',1)
p53_GOF_attractors = getAttractors(p53_GOF, type= c('asynchronous'))
plotAttractors(p53_GOF_attractors)

# To calculate basins of attraction we obtained attractors 
# using a synchronous update scheme. 

synchronous_attractors = getAttractors(discrete_network)
print(synchronous_attractors)
```
